```python
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/ML data Sets/fetal_health.csv')
data
```

```python
data.info()
```

```python
data=data.drop_duplicates()
data[data.duplicated()]
data
```

```python
#importing required libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
input_features=data.drop(['fetal_health'],axis=1)
print(input_features)
output_features=data['fetal_health']
output_features
```

```python
training_features,testing_features,training_test,testing_test=train_test_split(input_features,output_features,test_size=0.4,random_state=42)
clasifier=DecisionTreeClassifier()
classifier=clasifier.fit(training_features,training_test)
predictions=classifier.predict(testing_features)
score=accuracy_score(predictions,testing_test)
print("Accuracy Score : ",score)
```
