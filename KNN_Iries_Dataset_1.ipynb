```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/drive/MyDrive/ML data Sets/iris.csv')
data.head()
```

```python
data=data.drop_duplicates()
data[data.duplicated()]
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
data['variety']=label.fit_transform(data['variety'])
data
```

```python
from sklearn.metrics import accuracy_score
input_features=data.drop(['variety'],axis=1)
print(input_features)
output_features=data['variety']
print(output_features)
```

```python
from sklearn.model_selection import train_test_split
x,X,y,Y=train_test_split(input_features,output_features,test_size=0.4,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier as KNN
scaler=StandardScaler()
x=scaler.fit_transform(x)
X=scaler.fit_transform(X)
```

```python
classifier=KNN(n_neighbors=5,metric='euclidean')
classifier=classifier.fit(x,y)
predictions=classifier.predict(X)
score=accuracy_score(predictions,Y)
print("Accuracy Score : ",score)
```
