```python
import pandas as pd
data=pd.read_csv('/content/diabetes (1).csv')
```

```python
data.head(5)
```

```python
from sklearn.preprocessing import *
input_col=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
data_up=pd.get_dummies(data,columns=input_col)
data_up
```

```python
label=LabelEncoder()
data_up['Outcome']=label.fit_transform(data_up['Outcome'])
data_up
```

```python
x=data_up.drop(['Outcome'],axis=1)
y=data_up['Outcome']
from sklearn.model_selection import *
X_train,x_test,Y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
X_train
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(X_train,Y_train)
```

```python
y_predict=gnb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(classification_report(y_test,y_predict))
print("Accuracy Score :",accuracy_score(y_test,y_predict))
print("\nConfution Matris : ")
print(confusion_matrix(y_test,y_predict))
```
