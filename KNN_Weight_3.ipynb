```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/drive/MyDrive/ML data Sets/weight_height_bmi_dataset.csv')
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
data['Category']=label.fit_transform(data['Category'])
data
```

```python
from sklearn.metrics import accuracy_score
input_features=data.drop(['Category'],axis=1)
print(input_features)
output_features=data['Category']
print(output_features)
```

```python
from sklearn.model_selection import train_test_split
x,X,y,Y=train_test_split(input_features,output_features,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier as KNN
scaler=StandardScaler()
x=scaler.fit_transform(x)
X=scaler.fit_transform(X)
```

```python
classifier=KNN(n_neighbors=5,metric='euclidean')
classifier=classifier.fit(x,y)
predictions=classifier.predict(X)
score=accuracy_score(predictions,Y)
print("Accuracy Score : ",score)
```
